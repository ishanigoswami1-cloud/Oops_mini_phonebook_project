```python
class student:
    ''' this is class student to manage student safe and active 
    '''
    college_name="sipi college"
    departments=["Arts","commerce","science"]
    def greet(cls):
        print(cls)
        print(f"welcome to {cls.college_name}")
    def department(cls):
        print(cls)
        print(f"department in {cls.college_name} are:")
        for department in cls.departments:
            print(department)
    def study(self,n_hours): 
        print(f"self is:{self}")
        print(f"The student studies for {n_hours} hours a day!")
    def hobbies(self,hobbies_name):
        print(f" The student hobbies:{hobbies_name}")
    def __init__(self,name,roll):
        print(f"calling the initializer!:{self}")
        self.name= name 
        self.roll=roll

student1=student("rockey",34 )
print(f"The object : {student1}")
student1.study(2)
student1.hobbies("cricket")

"""
When we call an instance method using the object/instance of the class , python passes the object itself
as the first argument to that method 
that first argument
"""
student2=student("ronit",45)
print(f"The object : {student2}")
student2.study(3)
student2.hobbies("hockey")

student3=student("ronit",45)
print(f"The object : {student3}")
student3.study(8)
student3.hobbies("lorel")


print(student1.name, student1.roll)
print(student2.name, student2.roll)
print(student3.name, student3.roll)
print(student1.__dict__)
print(student2.__dict__)
print(student3.__dict__)
help(student)
print(student1.department)
print(student1.college_name)
print(student2.department)
print(student2.college_name)
## class Methods 
'''
class method are defines inside the class that are bound to the class 
to create a class method,we use a decorator-> classmethod 
'''
student1=student("rockey",34)
student1.study(4)
student1.greet()
student1.hobbies("cricket")
student1.department()
      
```

```python
##static method
'''
methods defined inside a class which is neither bound to the object nor to the class 
to create a static method , we use 
'''


class student:
    ''' this is class student to manage student safe and active 
    '''
    college_name="sipi college"
    departments=["Arts","commerce","science"]
    def greet(cls):
        print(cls)
        print(f"welcome to {cls.college_name}")
    def department(cls):
        print(cls)
        print(f"department in {cls.college_name} are:")
        for department in cls.departments:
            print(department)
    def study(self,n_hours): 
        print(f"self is:{self}")
        print(f"{self.name} studies for {n_hours} hours a day!")
    def hobbies(self,hobbies_name):
        print(f" the student plays{hobbies_name}in the college{self.college_name}")
    def __init__(self,name,roll):
        print(f"calling the initializer!:{self}")
        self.name= name 
        self.roll=roll
student1=student("rockey",34)
student1.study(4)
student1.greet()
student1.hobbies("cricket")
student1.department()
```

```python
class phonebook:
    phone_directory=[]
    def __init__(self,name,phone_number):
        self.name=name
        self.phone=phone_number
        phonebook.phone_directory.append(self)
    def show_contact(self):
        return f"Name:{self.name}, contact number:{self.phone}"

    @classmethod
    def show_all_contact(cls):
        if len(cls.phone_directory)==0:
            print("No contact found in the directory")
        else:
             for contact in cls.phone_directory:
                 print("contact.show_contact()")
    @classmethod
    def search_phonebook(cls,search_name):
        for contact in cls.phone_directory:
            if contact.name==search_name:
                return contact.phone
        print(f"No contact found for {search_name}")

    @staticmethod
    def validate_phone_number(phone_number):
        if len(phone_number)>=8 and phone_number.isdigit():
            return True
        else:
            return False
    n_phonebook=int(input("How many contacts do you want to add?: "))  
    for i in range(n_phonebook):
        name = input("Enter the name of the contact: ")
        phone_number = input("Enter the phone number: ")

        if validate_phone_number(phone_number):
            phonebook(name, phone_number)
            print("Contact saved successfully")

phonebook.show_all_contact()
#p1=phonebook("carol",234567899)
#p2=phonebook("rotus",987654321)
#print(phonebook.phone_directory)
#print(p1.show_contact())
#print(p2.show_contact())  

#print(phonebook.search_phonebook("carol"))
#print(phonebook.search_phonebook("john"))
```
